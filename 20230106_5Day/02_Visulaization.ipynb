```python
# 각 국가별 입국객 그래프 그리기 -> 반복문으로 이용해서 그리기 
## *주피터 리부팅 후 재 실행 방법 -> kernel > restart&RUN all
## 제목 : 각 국가의 이름이 들어가게 
# 자동화를 할 때는 변수를 많이 쓰면 좋다 


# 상위 5개 국가 뽑기 

temp = df_1 = df.pivot_table(values = "전체비율(%)",
              index = "국적",
              aggfunc = "mean").sort_values("전체비율(%)", ascending = False ).head(5)
temp.reset_index(inplace = True)

#상위 5개 국가 국적 정보 가져오기 : 튜플(옆으로), 시리즈(위에서 밑으로)형태 
## -> 리스트 처럼 사용하면된다  
## 시리즈 형태 ; ("중국", "일본", "대만", "미국", "홍콩")
name = temp["국적"]

for i in range (0,5,1) :
    # 각 국가 필터링하기 
    df_filter = df[df["국적"] == name[i]]
    
    # 그래프 그리기 
    plt.figure(figsize = (12,4))
    plt.plot(df_filter["기준년월"], df_filter["관광"])
    plt.title("{} 국적의 관광객 추이".format(name[i]))
    plt.xlabel("기준년월")
    plt.ylabel("관광")
    plt.xticks(["2010-01","2011-01", "2012-01", "2013-01", "2014-01", "2015-01",\
               "2016-01", "2017-01", "2018-01", "2019-01"])
```

```python
# 해답 
cntry_list = temp["국적"]

# 시리즈의 값을 이용하는 방식 
for cntry in cntry_list : 
    # 각 국가 필터링하기 
    df_filter = df[df["국적"] == cntry]
    
    # 그래프 그리기 
    plt.figure(figsize = (12,4))
    plt.plot(df_filter["기준년월"], df_filter["관광"])
    plt.title("{} 국적의 관광객 추이".format(cntry))
    plt.xlabel("기준년월")
    plt.ylabel("관광")
    plt.xticks(["2010-01","2011-01", "2012-01", "2013-01", "2014-01", "2015-01",\
               "2016-01", "2017-01", "2018-01", "2019-01"])
# 그래프 해석 
# 흐름이 비슷 : 이슈가 많이 없다 
# 중국 : 오르고 내리는 것은 유사 -> 2014-01부근에서 흐름이 변함 
#  2017-01 부근에서는 점진적으로 복구
# 일본 : 2015 중순즘 줄었다가 복귀
# 2015 -> 중국도 줄음, 복구 x
# 대만 : 2015 확 줄음 , 복구O 
# 미국 : 거의 비슷
# 홍콩 : 2015 중슨 즘 확 줄음 , 복구O
# => 2015 년쯤에 아시아권에 이슈가 있었을 것이다 
# 자세히 보기 위해 년도, 월별분석 -> 히트맵
## 기준년월의 년월 분리 해서 칼럼 만들기 
## 프로젝트 : 주제 정하기 -> 소주제 정하기(각 그래프명)
```

# 년도 및 월별 입국객수의 추이 분석

```python
# 년도 및 월에 대한 컬럼 추가하기 
# [문제] 기준년도에서 년도 월을 각각 추출해서 
## df 데이터 프레임에 "년도", "월" 칼럼을 추가하기 

# 데이터프레임에서 기준년월 가져오기

# 엄청 오래 걸린다 
# yymm_split = df["기준년월"].str.split('-')
# df["년"] = yymm_split.str.get(0)
# df["월"] = yymm_split.str.get(1)

# df
```

```python
#년월 추출 (방법-1)
# 리스트로 만들어서 넣기 
# 훨씬 빠르다 

yy = []
mm = []

for i in range(0,len(df),1) :
    yy.append(df["기준년월"][i][0:4])
    mm.append(df["기준년월"][i][5:7])

df["년"] = yy
df["월"] = mm
df
```

```python
# 년월추출 (방법-2)
# 이것도 빠름 

df["년"] = df["기준년월"].str.slice(0,4)
df["월"] = df["기준년월"].str.slice(5,7)
df
```

```python
# "국적"이 중국인 데이터만 조회하기 

df_filter = df[df["국적"] == "중국"]
df_filter 
```

```python
# 히트맵 그리기 : 3개 값(연도, 월, 입국객수)이 들어 감 
## 히트맵에 맞게 데이터 형태를 만들어 줘야 한다 : pivot_table
## 인덱스(index) : 년도
## 컬럼(columns) : 월 
## 데이터(values) : 관광 
## 사용하는 함수 속성 제시안함 
## 피봇 변수명 : df_pivot

df_pivot = df_filter.pivot_table(index = "년",
                          columns = "월",
                          values = "관광",
                          aggfunc = "mean")
# 데이터 프레임 형태가 아니라 피봇테이블 형태이다 

df_pivot
```

## 히트맵 시각화

```python
# 히트맵 라이브러리 불러들이기
# sseacorn이 matplotlib 기반이으로 항상 세트로 임포트 시키기
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 그래프 크기 지정 너미 16, 높이 10
plt.figure(figsize = (16,10))
# 2. 히트맵 그래프 함수에 데이터 넣기
## annot : 숫자값 표시여부
## fmt = ".0f" : float 타입 사용할 거고 소수점 있으면 표시
## cmap : 여러 옵션이 있음 
sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap="rocket_r")
# 3. 그래프 제목 넣기 
plt.title("년도 및 월별 관광객 추이")
# 4. 그래프 보여주기
plt.show()

## 그래프 해석 
# 본래 : 흐림 -> 진함 -> 흐림 
# 2015 : 진함 -> 흐림 -> 진함 => 2015년에 이슈가 있을 것이다 




```

```python
# top5 국가 각가에 대한 히트맵 그리기 

# 상위 5개 국가 뽑기 

temp = df.pivot_table(values = "전체비율(%)",
                      index = "국적",
                      aggfunc = "mean").sort_values("전체비율(%)", ascending = False ).head(5)
temp.reset_index(inplace = True)

name = temp["국적"]

# 히트맵 그리기 
for contry in name : 
    
    # pivot 하기 
    df_filter = df[df["국적"] == contry]
    
    df_pivot = df_filter.pivot_table(index = "년",
                          columns = "월",
                          values = "관광",
                          aggfunc = "mean")
    # 히트맵 그리기
    plt.figure(figsize = (16,10))
    sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap="rocket_r")
    plt.title("{} 년도 및 월별 관광객 추이".format(contry))
    plt.show()
# 2015년 아시아권 관광 데이터 이상 => 메르스
# 2017년 중국 => 사드 보복 
# 구글 기간 검색으로 이슈 확인하기 
```
