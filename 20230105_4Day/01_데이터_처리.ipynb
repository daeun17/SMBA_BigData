```python
# 테스트 용 (완성 x)
yy = 2010
mm =1 
continentis_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙',\
                    '교포소계']
continents = ["아시아"]* 25 + ["아메리카"]* 5 + ["유럽"]*23 + ["오세아니아"]*3 + ["아프리카"] *2 +["기타대륙"] + ["교포"]

kto_yymm = pd.read_excel("./files_02/kto_{}{}.xlsx".format(yy,str(mm).zfill(2)),
                        header = 1,
                        skipfooter = 4,
                        usecols = "A:G")

kto_yymm["기준년월"] = "{}-{}".format(yy,str(mm).zfill(2))

kto_c = kto_yymm[kto_yymm["국적"].isin(continentis_list) == False]

kto_c = kto_c.copy()

kto_c["대륙"] = continents

kto_c["관광객비율(%)"] = round((kto_c["관광"]/kto_c["계"]) * 100, 4)
kto_c.sort_values(by="관광객비율(%)", ascending = True)


kto_c["전체비율(%)"] = round(kto_c["관광"]/sum(kto_c["관광"]) * 100) 
kto_c.sort_values(by="전체비율(%)", ascending = True)

df = kto_c
df
```

```python
#모든 파일을 위 처리내용을 반영해서 하나의 데이터프레임으로 통합하기
## 최종 데이터프레임 이름 = df 합니다
## 1. for문으로 전체 통합 잘 되는지 확인 후 처리 내용부분을 함수로 분리하기
# 테스트용 (완성 x)

def DF (start, end) :
    yy = 2010
    mm =1 
    continentis_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙',\
                    '교포소계']
    continents = ["아시아"]* 25 + ["아메리카"]* 5 + ["유럽"]*23 + ["오세아니아"]*3 + ["아프리카"] *2 +["기타대륙"] + ["교포"]

    kto_yymm = pd.read_excel("./files_02/kto_{}{}.xlsx".format(yy,str(mm).zfill(2)),
                        header = 1,
                        skipfooter = 4,
                        usecols = "A:G")

    kto_yymm["기준년월"] = "{}-{}".format(yy,str(mm).zfill(2))

    kto_c = kto_yymm[kto_yymm["국적"].isin(continentis_list) == False]

    kto_c = kto_c.copy()

    kto_c["대륙"] = continents

    kto_c["관광객비율(%)"] = round((kto_c["관광"]/kto_c["계"]) * 100, 4)
    kto_c.sort_values(by="관광객비율(%)", ascending = True)


    kto_c["전체비율(%)"] = round(kto_c["관광"]/sum(kto_c["관광"]) * 100) 
    kto_c.sort_values(by="전체비율(%)", ascending = True)

    df = kto_c
        
    for i in range(start + 1 ,end + 1,1) :
        mm += 1
        if (i) % 12 == start :
            mm = 1
            yy += 1
            
        kto_yymm = pd.read_excel("./files_02/kto_{}{}.xlsx".format(yy,str(mm).zfill(2)),
                            header = 1,
                            skipfooter = 4,
                            usecols = "A:G")

        kto_yymm["기준년월"] = "{}-{}".format(yy,str(mm).zfill(2))

        kto_c = kto_yymm[kto_yymm["국적"].isin(continentis_list) == False]

        kto_c = kto_c.copy()

        kto_c["대륙"] = continents

        kto_c["관광객비율(%)"] = round((kto_c["관광"]/kto_c["계"]) * 100, 4)
        kto_c.sort_values(by="관광객비율(%)", ascending = True)


        kto_c["전체비율(%)"] = round(kto_c["관광"]/sum(kto_c["관광"]) * 100) 
        kto_c.sort_values(by="전체비율(%)", ascending = True)

        df = pd.concat([df,kto_c], ignore_index = True)
        
    return df 
```

```python
### 함수 생성하기
def get_kto_dataFrame(yyyy, mm) :
    file_nm = "./files_02/kto_{}{}.xlsx".format(yyyy, str(mm).zfill(2))
    #         print(file_nm)
    kto_201901 = pd.read_excel(file_nm,
                          header = 1,
                          usecols = "A:G",
                          skipfooter = 4)

    ### 기준년월 컬럼 생성하기
    kto_201901["기준년월"] = "{}-{}".format(yyyy, str(mm).zfill(2))   

    ### 국적에서 대륙이름 제거하기(정제)
    continents_list = ['아시아주', '미주', '구주', '대양주', 
                   '아프리카주', '기타대륙', '교포소계']
    condition = kto_201901["국적"].isin(continents_list)
    kto_201901_country = kto_201901[condition == False]

    ### 인덱스 재정렬 하기
    kto_201901_country_newindex =kto_201901_country.reset_index(drop=True)

    ### 대륙 컬럼 추가하기
    continents = ["아시아"]*25 + ["아메리카"]*5 + ["유럽"]*23 + ["오세아니아"]*3 + \
                ["아프리카"]*2 + ["기타대륙"] + ["교포"]
    kto_201901_country_newindex["대륙"] = continents

    ### 각 국가별로 관광을 위해 들어오는 비율을 계산하기...
    kto_201901_country_newindex["관광객비율(%)"] = \
                            round(kto_201901_country_newindex["관광"] / \
                              kto_201901_country_newindex["계"] * 100, 4)

    ### 전체 국가대비 관광객비율 계산하기...
    # - 전체 관광객 합산하기 : sum() 함수 사용하기
    tourist_sum = sum(kto_201901_country_newindex["관광"])
    kto_201901_country_newindex["전체비율(%)"] = \
            round(kto_201901_country_newindex["관광"] / tourist_sum * 100 , 4)
    
    return kto_201901_country_newindex
```

```python
df = pd.DataFrame()

for yyyy in range(2010, 2020, 1) :
    for mm in range(1, 13, 1) :
        try :
            ### 함수 호출하기
            temp = get_kto_dataFrame(yyyy, mm)
            
            ### 데이터프레임 추가하가(통합)
            df = pd.concat([df, temp], ignore_index = True)
            
        except :
            pass

print(len(df))
df
```
