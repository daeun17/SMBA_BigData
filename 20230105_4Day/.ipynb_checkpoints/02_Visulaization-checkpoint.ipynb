```python
## [문제] 중국 국적 데이터만 조회하기 
# 변수명 : df_filter 사용 
# 조회결과 건수 출력 
# 조회결과 출력 

df_filter = df[df["국적"] == "중국"]

print(len(df_filter))
df_filter
```

```python
# 선 그래프 그리기 : 기준년월별 관광객 추이 
# x 축, y축 순서 
plt.plot(df_filter["기준년월"], df_filter["관광"])
plt.show()
```

```python
# 그래프 디자인 
# 그래프 크기 조정 
## 넓이, 높이 순 
plt.figure(figsize = (12,4))

# 그래프 내용 설정
## x 축 y 축 순 
plt.plot(df_filter["기준년월"], df_filter["관광"])

# 그래프 제목 넣기 
plt.title("중국 국적의 관광객 추이")

# x축 이름 넣기 
plt.xlabel("기준년월")

#y축 이름 넣기 
plt.ylabel("관광")

# x 축의 값 새로 정의하기 : 범위로 
# 실제 존재하는 값으로 해야 한다 
plt.xticks(["2010-01","2011-01", "2012-01", "2013-01", "2014-01", "2015-01",\
           "2016-01", "2017-01", "2018-01", "2019-01"])

# 그래프 보이기 : 없어도 보이긴 함 
plt.show()
```

```python
# [문제]
# 우리나라에 방문하는 전체 외국인 관광객 수가 가장 많은 국가 순으로 top5 조회하기 
#조회 결과는 temp 변수 사용 

# 국가별로 가장 높은 값 추출하기 
df_1 = df.pivot_table(values = "관광",
              index = "국적",
              aggfunc = "sum")
# 추출한 데이터 프레임 중에서 top5 찾기 
temp = df_1.sort_values("관광", ascending = False ).head(5)

# 풀이 
temp = df_1 = df.pivot_table(values = "전체비율(%)",
              index = "국적",
              aggfunc = "mean").sort_values("전체비율(%)", ascending = False ).head(5)

temp
```

```python
# 핌소 테이블의 형태를 데이터 프레임 형태로 변환하기 
# reset_index()를 사용하면 피못테이블을 데이터 프레임으로 변환가능 
## inplace = True : (인덱스를 순차적으로 재정렬 후) 메모리에 자동 저장함 
temp.reset_index(inplace = True)
```
