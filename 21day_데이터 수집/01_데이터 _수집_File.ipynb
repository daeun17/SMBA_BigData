## CSV 파일 사용하기 

```python
# bicycle.csv 파일읽어들이기 
# 데이터 프레임 변수 ;df
# 상위 5개만 출력해보기 
```

```python
import pandas as pd 
```

```python
# UTF error 나면 encoding 설정해주기 
df1 = pd.read_csv("./data/bicycle.csv", encoding="euc-kr")
df1.head()
```

## EXCEL 파일 불러들이기 

```python
# bicycle.xlsx 파일 읽어들이기 
# 데이터 프레임 변수 : df2
```

```python
df2 = pd.read_excel("./data/bicycle.xlsx")
df2.head()
```

## Json 파일 읽어들이기 

```python
# read.json 파일 읽어들이기 
# df3 변수 사용
```

```python
df3 = pd.read_json("./data/read.json")
df3.head()
```

## 데이터 프레임 csv 로 저장하기

```python
# data 폴더에 저장 
# 저장 파일명 bicycle_save.csv 
#저장할 데이터 프레임 : df1 

df1.to_csv("./data/bicycle_save.csv", index =False)
```
