```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

file_path = "c:/ChromeDriver_exe/chrome_109_driver.exe"
url = "http://www.melon.com/chart/index.htm"
driver = webdriver.Chrome(file_path)
driver.get(url)

titles = driver.find_elements(By.CSS_SELECTOR,"tr div.ellipsis.rank01 > span > a")[0:10]
singers = driver.find_elements(By.CSS_SELECTOR,"tr div.ellipsis.rank02 > a")[0:10]

# print(singers[0].text)
# 곡이름 | 가수이름 형식으로 10개 출력하기 
for i in range (0, len(title), 1) :
     print(titles[i].text, singers[i].text, sep = "|")

#  웹 크롤링 결과를 {no : 1, title : 노래제목, singer : 가수이름} 리스트 형태로 만들기 
melon_list = []
for i in range (0, len(title), 1 ) :
    title = titles[i].text
    singer = singers[i].text
    melon_list.append(dict(zip(['no','title','singer'],[i+1,title,singer])))
    
print(melon_list)

time.sleep(3)
driver.quit()
```
