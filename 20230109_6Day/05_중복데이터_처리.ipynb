```python
# bicycle_dup.csv 파일읽어들인 후 정보확인하기 
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("./data/bicycle_dup.csv", encoding="euc-kr")
df.info()
```

```python
# 전체 컬럼들이 다음행에서 중복이 있는지 확인하기 
df[df.duplicated()]
```

```python
# 컬럼단위로 중복체크하기 
# 첫번째 데이터만 사용하고 나머지는 삭제 

df[df["이용거리"].duplicated()]
```

```python
df[["이용거리","나이"]]
df.duplicated(["이용거리","나이"])
```

```python
# 중복데이터 처리방법 지정 : keep 속성 사용 
## first : 첫번째 중복데이터는 살리고(false), 아래 나머지 true
## last : 마지막중복데이터는 살리고(false), 위에 나머지 true
## 생략가능 : 디폴트 값
# 가장 많이 사용하는 방식 
df.duplicated(["이용거리"],keep ="first")
df.duplicated(["이용거리"],keep ="last")
```

```python
# 중복데이터 삭제하기 
# keep = "first" : 첫번째 중복은 살리고, 나머지 뒤에 중복은 삭제
df.drop_duplicates(["이용거리"], keep ="first")
```

```python
# keep = False : 중복이 있는 데이터 모두 삭제 
df.drop_duplicates(["이용거리"], keep = False)
```
