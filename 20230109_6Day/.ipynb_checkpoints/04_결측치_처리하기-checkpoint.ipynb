# 결측치 처리 

```python
# 전처리 : 결측치 데이터, 중복 데이터, 이상치 데이터 처리 
# 결측치 : 특정 컬럼에 값이 없느 경우 어떻게 처리할 것인지 
# 중복 : 행들의 값들이 중복된 경우 러떻게 처리할 것인지 
# 이상치 : 특정범위를 벗어나는 값들에 대해서 어떻게 처리할 것인지 
# ==> 일반적인 처리는 삭제, 값 채우기 방식을 주로 사용 한다 
# ==> 이상치 처리의 경우만, 이상치 산출 계산공식을 사용하여 처리 

## 삭제시 행단위? 열 단위?  ,채우기는 빈도? 평균?, 주변 값? 
## 결측치 
### 분석시 데이터 결측치가 적다 -> 삭제 하는 편 (100개 중 1~2개 정도)
### 애매하다 100개중 10개 정도 -> 평균으로 채우기 
### 결측치가 많다 -> 비율로 채우기 (대신 정확하지 않다 데이터 적다 알리기)
## 중복 데이터 
### 중복 데이터는 삭제 -> 중복 데이터알린 후 진행 
## 이상치 : 데이터를 봤을 때 상식선을 벗어남, 비율 대비 벗어나는 데이터 
### -> 계산으로 찾아낸다 => 일반적으로는 제거, 
### 또는 수정(주변 컬럼데이터를 통해 확인 후 수정)
```

```python
# bicycle.csv 파일 읽어 들인 후  조회해주세요 
## 데이터프레임 변수명 : df 사용
```

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("./data/bicycle - 복사본.csv")
# error : utf-8 error, UnicodeDecodeError   
## 읽어 들이는 파일의 포멧이 안 맞다 
### 파일 메모장으로 열기 → 파일 → 다른이름으로 저장 → 인코딩 utf-8로 변경 
### 한글인 경우 아래처럼 포멧 타입 지정하면 잘 읽어 들여 진다 
#encoding : 수집데이터가 utf-8 포멧이 아닌경우 오류 발생
#        :한글데이터가 주로 오류 발생함 : euc-kr 설정 
df = pd.read_csv("./data/bicycle.csv", encoding="euc-kr")
df
```

```python
df.info()
# 대여거치대, 반납거치대, 이용시간, 이용거리 결측치 존재 
```

```python
# 결측치 확인하느 함수 : isnull()
## 있으면 True 없으면 False 
df.isnull()
```

```python
# 정상데이터만 확인하기 : notnull()
df.notnull()
```

```python
# 각 컬럼 별 결측치가 있는 개수 확인하기 
# sum(0) : 0의 의미는 행단위로 개수 sum 하기 
df.isnull().sum(0)
```

```python
# 행 단위별 결측데이터 칼럼의 개수 확인하기 
# sum(1) : 1의 의미는 컬럼의 개수를 의미함 
df.isnull().sum(1)
```

```python
df.info()
```

```python
# 결측치가 있는 행을 모두 삭제하기 
## 행단위 삭제 : axis = 0
df_dropallrow = df.dropna(axis=0)
df_dropallrow.info()
```

```python
# 결측치가 있는 컬럼 삭제하기 
## 컬럼 단위 삭제하기 : axis=1
## 보통 컬럼을 삭제하지는 않는다 
df_dropallcol = df.dropna(axis=1)
df_dropallcol.info()
```

```python
# 특정 컬럼을 기준으로 결측치가 있는 행 삭제하기 
df["대여소명"].dropna()
```

```python
df["대여거치대"].dropna()
# 동일 : df["대여거치대"].dropna(axis = 0)
```

```python
# 여러개 컬럼을 기준으로 행단위 삭제하기 
df[["대여소번호","대여거치대","이용시간"]].dropna(axis = 0)
```

### 결측치 처리 -> 채우기 

```python
### 결측치를 모두 0으로 채우기 (모든컬럼)
df.fillna(0).info()
```

```python
# 이용거리 컬럼의 결측치만 0으로 채워주세요 
df["이용거리"].fillna(0)
```

```python
# 이용거리 컬럼에 대한 결측치를 missing문자로 채우기 
df["이용거리"].fillna("missing")
```

```python
# 이용거리 컬럼의 결측 부분을 이용거리의 평균으로 채우기 
# 일반적으로 가장 많이 사용 
df["이용거리"].fillna(df["이용거리"].mean())
```
