## 데이터 통합하기 (행/열)

```python
s1 = pd.Series([0,1], index=["A","B"])
s2 = pd.Series([2,3,4],index= ["a","b","c"])
#행단위로 통합하기 : 기존과 동일
pd.concat([s1,s2],axis=0)
```

```python
df_temp1 = pd.DataFrame([["Dog",3],["Cat", 10],["Tiger",6],["Mouse",7]],
                        index = [0,1,2,3],
                       columns = ["동물","나이"])
df_temp1
```

```python
df_temp2 = pd.DataFrame([["집",0],["초원", 1],["수풀",0],["초원",1]],
                        index = [0,1,2,3],
                       columns = ["사는 곳","뿔 개수"])
df_temp2
```

```python
# merge : 컬럼간의 같은 컬럼이 존재 하는지 체크하여 통합 
# 길이가 동일하면 옆으로 통합 : concat(~,axis = 1)
# 컬럼 단위로 통합 : 조건 없음 
## 단, 행의 개수가 같은 경우 사용 

pd.concat([df_temp1, df_temp2], axis = 1)
```

```python
# merge 
df_info1 = pd.DataFrame({"고객번호" : [1001,1002,1003,1004,1005,1006,1007],
                        "이름" : ["강","김","홍", "이"," 빅", "최", "소"]},
                        columns = ["고객번호", "이름"])
df_info1
```

```python
df_info2 = pd.DataFrame({"고객번호" : [1001,1001,1005,1006,1008,1001],
                        "가격" : [1000,20000,15000,5000,100000,30000]},
                        columns = ["고객번호", "가격"])
df_info2
```

```python
# 데이터 옆으로 통합하기 (조건 들어감)
## inner : 두개의 데이터프레임에서 같은 조건 에 대해서만 가져오기 
df_merge = pd.merge(left = df_info1,
                   right = df_info2, 
                   how = 'inner')
df_merge
```

```python
## 데이터 프레임의 왼쪽을 기준으로 모든데이터 가져오고, 
## 오른쪽을 기준으로 모든 데이터 가져오기 
## outer : 데이터가 같든 같지 않든 가져오기 
df_merge = pd.merge(df_info1, df_info2, how = 'outer')
df_merge
```

```python
## left : 왼쪽 기준으로 가져오기 
df_merge = pd.merge(df_info1, df_info2, how = 'left')
df_merge
```

```python
# right : 오른쪽을 기준으로 가져오기 
df_merge = pd.merge(df_info1, df_info2, how = 'right')
df_merge
```
