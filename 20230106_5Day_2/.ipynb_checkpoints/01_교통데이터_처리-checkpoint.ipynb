```python
# 하차시각 캌럼 생성 
def YMH_H (df):
    yymmdd = []
    yymm = []
    dd = []
    hhmmss = []
    hh = []
    mm = []
    
    for i in range(0,len(df),1) : 
        yymmdd.append(str(df["하차시각"][i])[0:8])
        yymm.append(str(df["하차시각"][i])[0:6])
        dd.append(str(df["하차시각"][i])[6:8])
        
        hhmmss.append(str(df["하차시각"][i])[8:13])
        hh.append(str(df["하차시각"][i])[8:10])
        mm.append(str(df["하차시각"][i])[10:12])
    
    df["하차전체일자"] = yymmdd
    df["하차년월"] = yymm
    df["하차일"] = dd
    df["하차전체시간"] = hhmmss
    df["하차시간"] = hh
    df["하차분"] = mm
    
    
    return df 
```

```python
def YMH_S (df):
    yymmdd = []
    yymm = []
    dd = []
    hhmmss = []
    hh = []
    mm = []
    
    for i in range(0,len(df),1) : 
        yymmdd.append(str(df["승차시각"][i])[0:8])
        yymm.append(str(df["승차시각"][i])[0:6])
        dd.append(str(df["승차시각"][i])[6:8])
        
        hhmmss.append(str(df["승차시각"][i])[8:13])
        hh.append(str(df["승차시각"][i])[8:10])
        mm.append(str(df["승차시각"][i])[10:12])
    
    df["승차전체일자"] = yymmdd
    df["승차년월"] = yymm
    df["승차일"] = dd
    df["승차전체시간"] = hhmmss
    df["승차시간"] = hh
    df["승차분"] = mm
    
    return df 


    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 시각화 폰트 라이브러리 불러들이기 
from matplotlib import font_manager, rc

# 운영체제 확인하는 라이브러리 
import platform
plt.rc("font", family = "Malgun Gothic")
```

```python
#기준월 및 기준일자별 버스 이용량 분석 
df_pivot = df_merge2.pivot_table(index = "승차년월",
                          columns = "승차일",
                          values = "일별이용객수",
                          aggfunc = "mean")

plt.figure(figsize = (16,10))
sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap="rocket_r")
plt.title("월별 및 일별 이용객 추이")
plt.show()
```

```python
# 기준일 및 시간대별 버스 이용량 분석

# 월별 분리하기 
month = df_merge2.pivot_table(values = "시간별이용객수",
                              index = "승차년월",
                              aggfunc = "mean")
month.reset_index(inplace = True)
name = month["승차년월"]

for month in name :
    df_month = df_merge2[df_merge2["승차년월"] == month]
    
    df_pivot = df_month.pivot_table(index = "승차일",
                          columns = "승차시간",
                          values = "시간별이용객수",
                          aggfunc = "mean")

    plt.figure(figsize = (16,10))
    sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap="rocket_r")
    plt.title("{}월 일별 및 시간별 이용객 추이".format(month))
    plt.show()
```
