```python
"""
[주성분분석]
 - 3차원을 2차원으로
 - 특성이 많으면 훈련모델의 성능은 높아지지만, 과대적합 발생 가능
[주성분 분석 방법]
 - 데이터를 가장 잘 나타내는 일부 특성을 선택
 - 데이터 크기를 줄이면서, 학습 모델의 성능 향상하는 방법
[장점]
 - 대량의 데이터를 이용하여 훈련 시 PCA 방법을 이용하면,
   훈련 시간을 줄일 수 있음
"""
```

```python
### 사용할 비정형 데이터 읽어 들이기
import numpy as np
import matplotlib.pyplot as plt
# - *.npy : 넘파이 배열의 기본 저장 포멧
#         : 읽어 들일때는 넘파이의 load()함수 사용해야함
# - 데이터의 성격 : 3가지 과일에 대한 이미지 데이터
# - 사과, 바나나, 파인애플에 대한 이미지 데이터
file_path = "./data/09_fruits_300.npy"
fruits = np.load(file_path)
print(fruits.shape)

fruits_2d = fruits.reshape(-1, 100 * 100)
print(fruits_2d.shape)
"""
3차원 데이터 : (300, 100, 100)
 - 첫번째 값 : 행의 갯수 300개 행
 - 두번째 값 : 열의 갯수 100개 컬럼
 - 세번째 값 : 픽셀값
"""
```

## 주성분분석(PCA) 

```python
from sklearn.decomposition import PCA
### 모델 생성
# n_components : 축소하고자하는 주성분의 갯수 지정
pca = PCA(n_components=50)
### 훈련 시키기
pca.fit(fruits_2d)
```

```python
### 주성분 차원 확인하기
pca.components_.shape
```

```python
# arr : 3차원 데이터
# ratio(비율)
def draw_fruits(arr, ratio=1) :
    # 샘플의 갯수
    n = len(arr) 
    # 그림을 그릴 행의 갯수 지정 : 10으로 나누어서 행의 갯수 결정
    # 한줄에 10개씩 이미지 그리기위해서
    rows = int(np.ceil(n/10))
    # 행이 1개일 때 : 열 개수는 샘플 개수, 그렇지 않으면 10개
    cols = n if rows < 2 else 10
    ### subplot 설정
    # squeeze : 2차원 형태로 axs를 반환할 수있도록 함
    #         : subplot의 공간을 행렬의 2차원으로 만들기 위해서..
    fig, axs = plt.subplots(rows, cols,
                            figsize=(cols*ratio, rows*ratio),
                            squeeze=False)
    for i in range(rows) :
        for j in range(cols) :
            # n개 까지만 그리기
            if (i * 10 + j) < n:
                axs[i, j].imshow(arr[i*10+j], cmap="gray_r")
            axs[i, j].axis("off")
    plt.show()
```

```python
### 주성분으로 훈련한 결과 시각화 하기
draw_fruits(pca.components_.reshape(-1, 100, 100))
```

```python
### PCA를 이용한 주성분 차원축소(특성의 갯수 줄이기)
fruits_pca = pca.transform(fruits_2d)
fruits_pca.shape
```

```python
### 원본데이터를 주성분으로 재구성하기 : 이미지 다시 복원(주성분으로만..)
fruits_inverse = pca.inverse_transform(fruits_pca)
fruits_inverse.shape
```

```python
### 3차원 데이터로 변환하기 : 그림 그리기 위해서
fruits_reconstruct = fruits_inverse.reshape(-1, 100, 100)
fruits_reconstruct.shape
```

```python
### 시각화 하기
for start in [0, 100, 200] :
    draw_fruits(fruits_reconstruct[start : start + 100])
    print("\n")
```

```python
### 주성분의 갯수 확인하기
plt.plot(pca.explained_variance_ratio_)
plt.xlabel("PCA count")
# y값은 분산값
plt.ylabel("Valiance Ratio")
plt.show()
```

## 다른 모델에 적용해 보기 

### 로지스틱회귀 모델에 적용

```python
from sklearn.linear_model import LogisticRegression

## 모델 생성 
lr = LogisticRegression()
lr
# 정형 데이터를 사용 하고, 지도학습을 한다 
# 종속 변수가 필요 -> target data 가 필요 
```

```python
# 로지스틱 회귀 모델은 정형데이터를 사용하는 지도학습 모델이다 
# - 종속변수(target) 가 필요함 
# - 사용하는 데이터는 3개의 그룹 : 0,1,2의 그룹으로 나눠져 있고 
# 이 그룹 값이 종속변수값이 된다 
# 각 그룹은 100 개씩...
# 변수명 target으로 하는 종속변수를 만들어 주세요 
# target = ["0" * 100 + "1" * 100 + "2" * 100 ]
# target
target = np.array([0]*100 + [1]*100 + [2]*100)
target.shape
```

```python
# 현재 fruits 데이터 형태 보기 
import pandas as pd 
df = pd.DataFrame(fruits_2d)
df.head(5)
```

```python
# 교차검증으로 로지스틱회귀 모델 훈련 정확도 확인해보기 
from sklearn.model_selection import cross_validate
# 주성분 분석(PCA)을 하지 않은 fruits_2d 데이터 사용 
scores = cross_validate(lr,fruits_2d, target)
print(scores)

print(round(np.mean(scores['test_score']),3))
print(round(np.mean(scores["fit_time"]),3))
print(round(np.mean(scores['score_time']),3))
```

```python
# 주성분분석을 한 데이터 사용 
scores = cross_validate(lr,fruits_pca, target)
print(scores)

print(round(np.mean(scores['test_score']),3))
print(round(np.mean(scores["fit_time"]),3))
print(round(np.mean(scores['score_time']),3))
# 시간도 줄고 정확도도 올라갔다 
```

```python
# 개수의 패턴이 정해져 있는 경우 -> 로지스틱(분류)
# 패턴이 일정하지 않는 경우 -> 군집 사용 
```

### KMeans(K-평균)에 PCA 적용하기 

```python
# KMeans 모델에 적용하기 
# 클러스터 개수 어제 결정된 값 사용 
# 모델 변수 : km 
#  훈련시 사용 데이터 : pca 데이터 
from sklearn.cluster import KMeans

# 모델 생성하기 
km = KMeans(n_clusters=3, random_state=42)

#훈련시키기 
km.fit(fruits_pca)
```

```python
# 그룹을 몇개로 했는지 확인 
print(np.unique(km.labels_, return_counts=True))
# 100 개를 넘어가거나 적으면 그 만큼 못 맟춘 것
```

```python
# 시각화 
for label in range(0,3) :
    draw_fruits(fruits[km.labels_ == label])
    print("\n")
```

```python
# 산점도로 시각화하여 3개의 그룹의 분포 확인하기 
for lable in range(0,3) :
    data = fruits_pca[km.labels_ == label]
    plt.scatter(data[:],data[:2])
plt.legend(['apple','banana','pineapple'])
plt.show()
```
